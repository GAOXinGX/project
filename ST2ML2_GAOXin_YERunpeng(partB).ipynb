```python
import pandas as pd
import os
#import unicodedata
```

### Create fuctions to get positive and negative scores by using vader

```python
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
analyser = SentimentIntensityAnalyzer()

def getpos(text):
    a = analyser.polarity_scores(text)
    return a['pos']

def getneg(text):
    a = analyser.polarity_scores(text)
    return a['neg']
```

### Extract imformation and process data into clean dataframe

```python
text = []
label = []
prezd = []
score_pos = []
score_neg = []

path = r"C:/Users/that/OneDrive - Efrei/s8/ST2ML2 - Advanced Machine Learning and Text Mining/TP project1/presidents-speeches"
president_names = os.listdir(path)

for president in president_names:
    speeches = os.listdir(path + "/" + president + "/speeches")
    num_speeches = len(speeches)
    i=1
    for speech in speeches:
        f = open(path + "/" +president + "/speeches/" + speech, "r")
        content = f.read()
        pos = getpos(content)
        neg = getneg(content)
        
        text.append(content)
        prezd.append(president)
        label.append(speech)
        score_pos.append(pos)
        score_neg.append(neg)
        
def delete(s):
    return s[2:]

text=list(map(delete,text))

def delete(s):
    return s[:-4] 

label=list(map(delete,label))

def split(s):
    return s.split("__")

label=list(map(split,label))
```

```python
df = pd.DataFrame()
df['text'] = text
df['president'] = prezd
df['label'] = label
df['score_pos'] = score_pos
df['score_neg'] = score_neg

df1=df['label'].apply(pd.Series,index=['date','speech'])
df = df.join(df1)
df = df.drop(labels='label',axis=1)
df['date'] = pd.to_datetime(df['date'])
df.sort_values(by=['date'], inplace=True, ascending=True)
df = df.reset_index(drop=True)
df
```
