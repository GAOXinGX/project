```python
from sklearn import metrics
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
A, b = df1[['AT', 'V', 'AP', 'RH']], df1['PE']
```

```python
A_train, A_test, b_train, b_test = train_test_split(A, b, test_size = 0.2)
```

```python
LR = LinearRegression()
LR.fit(A_train, b_train)
```

```python
b_train_pred = LR.predict(A_train)
b_test_pred = LR.predict(A_test)
b_train_mae = metrics.mean_absolute_error(b_train, b_train_pred)
b_test_mae = metrics.mean_absolute_error(b_test, b_test_pred)
print('train-MAE: ', b_train_mae)
print("test-MAE: ", b_test_mae)
```

```python
b_train_mse = metrics.mean_squared_error(b_train, b_train_pred)
b_test_mse = metrics.mean_squared_error(b_test, b_test_pred)
print('train-MSE: ', b_train_mse)
print("test-MSE: ", b_test_mse)
```

```python
b_train_r2 = metrics.r2_score(b_train, b_train_pred)
b_test_r2 = metrics.r2_score(b_test, b_test_pred)
print('train-R2: ', b_train_r2)
print("test-R2: ", b_test_r2)
```

```python
b_train_rmse = sqrt(b_train_mse)
b_test_rmse = sqrt(b_test_mse)
print('train-RMSE: ', b_train_rmse)
print("test-RMSE: ", b_test_rmse)
```
